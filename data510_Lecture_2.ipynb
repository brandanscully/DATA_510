__[argument]__ __[flags]__ __[directory]__

# Current Directory Path

The __Full Path__ or __Absolute Path__ describes the complete hierarchy from the __Root Directory__ to something.

The __Current Directory__ is the directory your terminal is pointed to.

To get the __path__ use pwd



The __Root Directory__ in this case is _/Users_

# Changing Directories

To __c__hange __d__irectories, use __cd__

to go to a directory, __cd path__



to go back one dircetory, __cd ..__

to go back two directories, __cd ../..__

to your home directory, __cd__

# Viewing the contents of a directory

to __l__ist files in a directory, use __ls__

to list them in descending order by size, add __-S__



# Creaty empty things

to __make__ a new empty __directory__, use __mkdir__ _name_

    % mkdir newFolder

to make a new empty file, use __touch__ _name_ (dot) _ext_

    % touch someFile.txt

# Display file content

to show file content, use __cat__ _name_

    % cat someFile.txt

to __enumerate__ file content, use __cat__ __-n__ _name_

    % cat -n someFile.txt

to show only the __head__ of a file to includ x lines, use __head__ __-n x__ _name_
   
    % head -n 3 someFile.text
    
to show only the __tail__ of a file to includ x lines, use __tail__ __-n x__ _name_
   
    % tail -n 3 someFile.text

# Moving or Renaming or Both

you can __move__ and/or __rename__ a file with __mv__

to __move__, use __mv__ _path_ _newPath_

    % mv someFile.txt /Users/brandanscully/Documents/Github/DATA_510

to __rename__, use __mv__ _name_ _newName_

    % mv someFile.txt myFile.txt

to __move__ _and_ __rename__, use __mv__ _path/name_ _newPath/newName_

    % mv someFile.txt /Users/brandanscully/Documents/Github/DATA_510/myFile.txt

# Removing files

to __remove__ a file, use __rm__ _path/name01_ *multiple files _path/name02_ _path/name03_

    % rm someFile.txt  (BE CAREFUL!)

# Version Control: Git & GitHub
## [Git/Github Experience](https://docs.google.com/forms/d/19is5DW07sVHBZZG4xdAjfGInpY6xQ3aqxZVzN-gQ7kM/edit#responses)

__Version control__ allows one or multiple users to keep track of edits to a file(s).

__[Git](https://git-scm.com)__ is an open source version control system.

* Very handy when you break things, because you can _rewind_ to earlier (working) versions.
* Makes file sharing less cumbersome and less confusing than emailing attachments.
* Helpful to track progress because comments describe what was done and by whom with each commit.
* Provides the ability to require code review before implementation

__[GitHub](https://github.com)__ is a popular implementation of Git.

# To Use Git

* Go to __[GitHub](https://github.com)__
* Sign up for GitHub with an email address.
* Download and install a Git GUI (I'm using _GitHub Desktop_)
* Here's a __[tutorial](https://docs.github.com/en/get-started/quickstart/hello-world)__ from GitHub.

# Terms to know

__Repository__ or __Repo__ is a folder with Git implemented.

To __clone__ a __repo__ means to make a copy.

__Branch__ is an alternative version of a _repo_.

__Commit__ is a saved change to a version-controlled file. Normally you will add a comment describing the change.

Here's a __[glossary](https://git-scm.com/docs/gitglossary)__ from Git.
